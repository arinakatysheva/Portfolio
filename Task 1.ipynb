## Задача 1 (1 балл)

Даны три числа a, b и c. Выполните обмен значений следующим образом:

- a становится равным сумме b и c;

- b становится равным разности a (исходного) и c;

- c становится равным произведению a и b (исходных).

_Пример:_

_In: `a = 2, b = 3, c = 4`_

_Out: `a = 7, b = -2, c = 6`_ 

```python
# здесь напишите решение (ᵔ◡ᵔ)

# Исходные значения
a = 2
b = 3
c = 4
print(a, b, c)

# Циклический обмен значений
a, b, c = (b + c), (a - c), (a * b)
print(a, b, c)
```

### Альтернативное решение через дополнительные переменные

_Пример 1:_

_In: `11`_

_Пример 2:_

Так как на первых двух занятиях еще не проходили циклы

### Решение через условный оператор if

```python
x = int(input('Введите число:'))

if x % 2 == 0:
    x = 'even'
else:
    x = 'odd'

print(x)
```

## Задача 5 (1 балл)

Проверьте, является ли пароль надёжным, критерии надежности:

- длина не меньше 8 символов;

- содержит хотя бы одну цифру,

- не начинается и не заканчивается пробелом.

Задача решается без использования условного оператора.

_Пример 1:_

_In: `secret123`_

_Out: `Пароль надёжный`

_Пример 2:_

_In: `123`_

_Out: `Пароль ненадёжный`_

```python
# здесь напишите решение (ᵔ◡ᵔ)

# Исходные значения
a = input('Введите пароль: ')

# Вывод результата
check = len(a) >= 8 and a.isalpha() == False and a == a.strip()
print({True: "Пароль надёжный", False: "Пароль ненадёжный"}[check])
```

### Проверка надежности пароля, если в нем содержатся специальные символы 
1. Оставляем в строке только те символы, которые подходят под условие isalnum(), то есть символ является цифрой или буквой.
2. Проверяем, что оставшаяся часть строки (без спецсимволов) содержит не только буквы, но и цифры. 

```python
a = input('Введите пароль: ')

a_new = "".join(ch for ch in a if ch.isalnum())

check = len(a) >= 8 and not a_new.isalpha() and a == a.strip()
print({True: "Пароль надёжный", False: "Пароль ненадёжный"}[check])
```

## Задача 6 (1 балл)

Необходимо рассчитать общий доход сотрудника, который получает базовую зарплату, премию за KPI, доплату за переработки и бонус за стаж. 

Входные данные:

- размер базовой зарплаты (БЗ),
- премия за KPI (KPI), 
- процент переработок (П), 
- количество лет стажа (С). 


<p style="text-align: center;">Формула расчета дохода:</p>

<p style="text-align: center;">Доход = (БЗ + KPI) * (1 + П / 100) + Бонус за стаж</p>

<p style="text-align: center;">Бонус за стаж = БЗ * 0.03 * С</p></p>


_In:_

_`БЗ: 60000`_

_`KPI: 12000`_

_`П: 15`_

_`С: 4`_

_Out:_

_`90000.0`_

```python
# здесь напишите решение (ᵔ◡ᵔ)

# Исходные значения
bz = float(input("БЗ: "))
kpi = float(input("KPI: "))
p = float(input("П: "))
s = float(input("С: "))

# Расчет бонуса за стаж
bonus = bz * 0.03 * s

# Расчет общего дохода
income = (bz + kpi) * (1 + p / 100) + bonus

print('Общий доход сотрудника =', income)
```

Ниже приведен пример, когда все навыки соотвествуют должности. То есть для выполнения условия нужно, чтобы в строке присуствовали все требуемые навыки. При этом порядок их перечисления неважен.

```python
position = input('Введите должность: ')
skills = input('Перечислите навыки: ')  

required_skills = (["Python", "SQL", "ML"] * (position.title() == "Data Scientist") or
                   ["SQL", "Excel", "Python"] * (position.title() == "Analyst") or
                   ["ML", "Python", "Docker"] * (position.title() == "Ml Engeneer"))

print(f"Соответствие навыков должности: {all(skill in skills for skill in required_skills)}")   # Равно True, если есть все требуемые навыки
```

Помимо булевого значения, вычислим `процент навыков`, которые соотвествуют требуемым, чтобы по этому показателю можно было сравнивать кандидатов. 

```python
match_percent = sum(skill in skills for skill in required_skills) / len(required_skills)
print(f"Совпадение навыков: {match_percent:.2%}")
```

Можно добавить `индикатор совпадения навыков`, чтобы HR было проще ранжировать кандидатов. 

```python
high_rate = 0.8
low_rate = 0.5  # Определяем условные границы интервалов 

indicator = ("Полностью соотвествует" * (match_percent >= high_rate) +
          "Частично соотвествует" * (low_rate <= match_percent < high_rate) +
          "Недостаточно навыков" * (match_percent < low_rate))

print(f"Совпадение навыков: {match_percent:.2%} - {indicator.lower()}")
```

За основу расчета размера `зарплаты` возьмем формулу из прошлого задания:
<p style="text-align: center;">Ежемесячный доход = Оклад + Бонус</p></p>
<p style="text-align: center;">Бонус за стаж = Оклад * 0.03 * Стаж</p></p>

```python
experience = 3  # Добавляем новые входные данные (стаж сотрудника)

salary = (100000 * (grade.title() == "Junior") +
          150000 * (grade.title() == "Middle") +
          200000 * (grade.title() == "Senior"))  # Размер оклада определяем исходя из грейда сотрудника

month_income = salary + (salary * 0.03 * experience)

print(f"Оклад: {salary:,}")
print(f"Ежемесячный доход: {month_income:,.2f}")
```

Чтобы посчитать `совокупный доход за год`, сначала определяем, сколько кварталов отработает сотрудник в этом году в зависимости от даты его найма. Затем вычисляем сумму дохода по следующей формуле:  
<p style="text-align: center;">Совокупный годовой доход = Ежемесячный доход * Отработано месяцев + Годовой бонус</p></p>
<p style="text-align: center;">Годовой бонус =  Оклад * 0.5 * 1.5 * KPI</p></p>
<p style="text-align: center;">KPI = Отработано кварталов / 4</p></p>

```python
month = int(date.split(".")[1])  # Забираем из даты найма месяц
quarter = (month - 1) // 3 + 1  # Определяем соответствующий месяцу квартал

kpi = (4 - quarter + 1) / 4
year_bonus = salary * 0.5 * 1.5 * kpi

year_income = month_income * (12 - month + 1) + year_bonus
print(f"Совокупный годовой доход: {year_income:,.2f}")
```

Можно рассчитать `доход сотрудника за 3 года`, чтобы компания отразила это в своем бизнес-плане. Так как на размер ежемесячного дохода влияет стаж сотрудника, то считаем доход за каждый год по-отдельности. 

```python
# 2-й год (стаж +1)
second_month_income = salary + (salary * 0.03 * (experience + 1))
second_year_income = second_month_income * 12 + (salary * 0.5 * 1.5 * 1)  # KPI будет равен 1 во 2-й и 3-й годы 

# 3-й год (стаж +2)
third_month_income = salary + (salary * 0.03 * (experience + 2))
third_year_income = third_month_income * 12 + (salary * 0.5 * 1.5 * 1)

print(f"Доход за 2-й год работы: {second_year_income:,.2f}")
print(f"Доход за 3-й год работы: {third_year_income:,.2f}")
print(f"\nСовокупный доход за 3 года: {(year_income + second_year_income + third_year_income):,.2f}")
```
