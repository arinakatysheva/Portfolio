# Задача 1 (1 балл)

Проанализируйте резюме кандидатов и определите, на какие позиции они подходят лучше всего. У нас есть словарь `position_requirements`, который содержит требования к ключевым навыкам для различных должностей: Data Scientist, Web Developer и DevOps Engineer. Также имеется словарь `resumes` с информацией о навыках каждого кандидата, взятых из резюме.

Задача заключается в том, чтобы для каждого резюме определить, каким позициям соответствуют навыки кандидата. Кандидат считается подходящим для позиции, если обладает всеми навыками, указанными в требованиях к этой позиции.

Необходимо написать код, который проанализирует данные по каждому резюме, сравнит навыки кандидата с требованиями каждой позиции и выведет список подходящих должностей для каждого кандидата. 

На основе приведенных ниже данных должен получиться следующий результат:

`resume_101.pdf: ['Data Scientist', 'Web Developer']`

`resume_102.pdf: ['Web Developer']`

`resume_103.pdf: ['DevOps Engineer']`

`resume_104.pdf: []`

```python
# Ключевые навыки для различных позиций
position_requirements = {
    "Data Scientist": frozenset(["Python", "ML", "Статистика", "SQL"]),
    "Web Developer": frozenset(["JavaScript", "React", "Node.js", "HTML/CSS"]),
    "DevOps Engineer": frozenset(["Docker", "Kubernetes", "AWS", "CI/CD"])
}

# Резюме кандидатов с их навыками
resumes = {
    "resume_101.pdf": frozenset(["Python", "ML", "Статистика", "SQL", "Deep Learning", "JavaScript", "React", "Node.js", "HTML/CSS"]),
    "resume_102.pdf": frozenset(["JavaScript", "React", "Node.js", "HTML/CSS", "Redux"]),
    "resume_103.pdf": frozenset(["Docker", "Kubernetes", "AWS", "Python", "Bash", "CI/CD"]),
    "resume_104.pdf": frozenset(["Python", "SQL", "Excel", "Tableau"])
}
```

```python
# здесь напишите решение

# Создаем пустой словарь 
matches = {}

for resume, skills in resumes.items():
    suitable_positions = []
    for position, required_skills in position_requirements.items():
        if required_skills.issubset(skills):  # проверяем наличие всех нужных навыков
            suitable_positions.append(position)
    matches[resume] = suitable_positions

# Вывод результата
for resume, positions in matches.items():
    print(f"{resume}: {positions}")
```

# Задача 2 (1 балл)

Напишите функцию `average(numbers)`, которая возвращает среднее значение списка чисел. Если подается пустой список, то она возвращает `Ошибка: список пуст`. Проверьте её работу с помощью `assert` на пустом и непустом списках.

_Пример:_

_In: `[1, 2, 3, 4, 5]`_

_Out: `3.0`_

---

_In: `[-5, 0, 5]`_


_Out: `0.0`_

---

_In: `[]`_

_Out: `Ошибка: список пуст`_


```python
# здесь напишите решение

def average(numbers) -> float:
    if not numbers:
        return "Ошибка: список пуст"  # выполняется, если подается пустой список 
    return sum(numbers) / len(numbers)  # выполняется, если подается непустой список 

# Проверка работы функции
assert average ([1, 2, 3, 4, 5]) == 3.0
assert average ([-5, 0, 5]) == 0.0
assert average ([]) == "Ошибка: список пуст"

print("Все проверки прошли успешно!")
```

# Задача 4 (1 балл)

Напишите функцию `check_password`, которая проверяет переданный ей пароль на сложность и выводит результат проверки. Сложным паролем считается комбинация символов, в которой :
- Есть хотя бы **3 цифры**;
- Есть хотя бы **одна заглавная буква**;
- Есть хотя бы **одна строчная буква**;
- Есть хотя бы **один символ** из следующего набора `!@#$%*`;
- Общая длина **не менее 10 символов**.

Если пароль прошел все проверки, функция должна вывести на экран фразу `Надежный пароль`, в противном случае - `Ненадежный пароль`.

_Пример:_

_In: `check_password('pAs333sw*ord')`_

_Out: `Надежный пароль`_

---

_In: `check_password('pas__ord')`_


_Out: `Ненадежный пароль`_


```python
# здесь напишите решение

def check_password(password):
    """Сначала проверяем общую длину пароля """
    if len(password) < 10:
        print("Ненадежный пароль")
        return

    """Определяем локальные переменные"""
    digits = 0
    upper = 0
    lower = 0
    special = 0
    specials = "!@#$%*"

    """Если длина подходит, проверяем другие характеристики пароля"""
    for passw in password:
        if passw.isdigit():
            digits += 1
        elif 'A' <= passw <= 'Z':
            upper += 1
        elif 'a' <= passw <= 'z':
            lower += 1
        elif passw in specials:
            special += 1

    """Проверяем одновременное выполнение всех условий"""
    if digits >= 3 and upper > 0 and lower > 0 and special > 0:
        print("Надежный пароль")
    else:
        print("Ненадежный пароль")
```

```python
check_password('pAs333sw*ord')
check_password('pas__ord')
```

# Задача 5 (2 балла)

В HR-отделе функционирует система, которая помогает подбирать кандидатов под вакансии. Нужно написать функцию `match_candidate`, которая будет сравнивать навыки кандидата с требованиями вакансии.

Функция принимает три аргумента:
- `candidate_name` — имя кандидата (строка).
- `candidate_skills` — список навыков кандидата (список строк).
- `job` — словарь с информацией о вакансии. У словаря два ключа:
    - `"title"` — название вакансии (строка),
    - `"skills"` — список необходимых навыков (список строк).

Функция возвращает кортеж из двух значений:
- булево значение (True или False) — подходит ли кандидат (если совпало больше половины навыков),
- число с плавающей точкой от 0.0 до 1.0 — доля совпавших навыков.

**!** Все параметры функции и возвращаемое значение должны быть снабжены аннотациями типов.

_Пример на основе приведенных ниже данных:_

_In: `match_candidate("Анна", ["Excel", "Word"], job)`_

_Out: `(False, 0.0)`_

---

_In: `match_candidate("Иван", ["Python", "Git"], job)`_


_Out: `(False, 0.5)`_

---

_In: `match_candidate("Мария", ["Python", "Django", "SQL", "Git"], job)`_

_Out: `(True, 1.0)`_


```python
job = {
    "title": "Python Developer",
    "skills": ["Python", "Django", "SQL", "Git"]
}
```

```python
# здесь напишите решение 

def match_candidate(candidate_name: str, candidate_skills: list[str], job: dict[str, list[str]]) -> tuple[bool, float]:
    required_skills = job["skills"]
    total = len(required_skills)

    """Подсчитываем количество совпавших навыков"""
    matched_skills = 0
    for skill in candidate_skills:
        if skill in required_skills:
            matched_skills += 1

    metric = matched_skills / total if total > 0 else 0.0
    """Кандидат подходит, если совпало больше половины навыков"""
    suitable_candidate = metric > 0.5  

    return (suitable_candidate, round(metric, 1))
```

```python
print(match_candidate("Анна", ["Excel", "Word"], job))
print(match_candidate("Иван", ["Python", "Git"], job))
print(match_candidate("Мария", ["Python", "Django", "SQL", "Git"], job))
```

# Задача 6 (2 балла)

HR-система получает данные о кандидатах из резюме в текстовом файле `resumes.txt`. 

О каждом кандидате в файле содержаться четыре записи:

- Имя: Иван Иванов
- Возраст: 29
- Навыки: Python, SQL, Data Analysis
- Опыт (лет): 5
---
- Имя: Анна Смирнова
- Возраст: 35
- Навыки: HR, Recruitment, Team Management
- Опыт (лет): 10

Между записями о кандидатах пустые строки.

Необходимо прочитать содержимое файла и для каждого кандидата сформировать словарь с ключами в следующем формате:

    {
        'Иван Иванов': {'Возраст': 29, 'Навыки': ['Python', 'SQL', 'Data Analysis'], 'Опыт (лет)': 5},
        'Анна Смирнова': {'Возраст': 35, 'Навыки': ['HR', 'Recruitment', 'Team Management'], 'Опыт (лет)': 10}
    }


Далее нужно выполнить следующее:
- Посчитать средний возраст и опыт кандидатов.
- Найти имена кандидатов, у которых в навыках встречается "Python".
- Записать эту информацию в новый файл.

```python
# здесь напишите решение

def check_resumes(input_file: str, output_file: str):
    candidates = {}

    """Читаем содержимое файла"""
    with open(input_file, "r", encoding="utf-8") as f:
        employee_data = f.read().strip()

    """Разделяем резюме по пустым строкам между ними"""
    resumes = employee_data.split("\n\n")

    for employee in resumes:
        lines = employee.strip().split("\n")
        data = {}
        name = None

        for line in lines:
            if line.startswith("Имя:"):
                name = line.split(":", 1)[1].strip()
            elif line.startswith("Возраст:"):
                data["Возраст"] = int(line.split(":", 1)[1].strip())
            elif line.startswith("Навыки:"):
                skills_str = line.split(":", 1)[1].strip()
                data["Навыки"] = [skill.strip() for skill in skills_str.split(",")]
            elif line.startswith("Опыт (лет):"):
                data["Опыт (лет)"] = int(line.split(":", 1)[1].strip())

        if name:
            candidates[name] = data

    """Считаем средний возраст"""
    avg_age = sum(c["Возраст"] for c in candidates.values()) / len(candidates)
    
    """Считаем опыт кандидатов"""
    avg_experience = sum(c["Опыт (лет)"] for c in candidates.values()) / len(candidates)

    """Находим кандидатов со знанием Python"""
    python_candidates = [name for name, c in candidates.items() if "Python" in c["Навыки"]]

    """Соединяем все для записи в новый файл"""
    final_table = [
        f"Средний возраст кандидатов: {avg_age:.1f}",
        f"Средний опыт (лет): {avg_experience:.1f}",
        f"Кандидаты с навыком Python: {', '.join(python_candidates) if python_candidates else 'Кандидаты отсутствуют'}\n",
    ]

    for name, info in candidates.items():
        final_table.append(f"{name}: {info}")

    """Записываем результат в новый файл"""
    with open(output_file, "w", encoding="utf-8") as f:
        f.write("\n".join(final_table))
```

## Дополнительное задание
HR-департамент компании "Альфа" планирует провести корпоративное обучение для сотрудников. У каждого курса есть список необходимых для его изучения навыков, а у каждого сотрудника — их навыки и уровень мотивации (от 1 до 10).

Необходимо написать программу для подбора сотрудников, чтобы определить кто лучше всего подходит для обучения по каждому курсу на основе совпадения навыков и высокого уровня мотивации.

Данные о сотрудниках и курсах хранятся в текстовых файлах `employees.txt` и `courses.txt`.
1. Проверить файлы на корректность данных.
2. Удалить пустые строки и пробелы в данных. 
3. Преобразовать данные из файлов в словари. 
4. Отобрать кандидатов для каждого курса (кандидат подходит, если доля совпадающих навыков и мотивация превышают пороговое значение).
5. Итоговый отчет записать в новый файл. 

```python
def hr_training(*args, **kwargs):
    motivation_limit = kwargs.get("motivation_limit", 6)  # минимальный уровень мотивации
    min_match = kwargs.get("min_match", 0.5)  # порог совпадения навыков

    """Исходные данные по сотрудникам и курсам"""
    employees_data = [
        "Александр|Excel,SQL,Git,Time management|7",
        "Мария|Python,ML,Git,Leadership|9",
        "Юлия|Stata,Excel,Communication|3",
        "Евгений|Docker,Python,SQL|8",
        "Петр|A/B tests,Python,ML|7",
        "Виолетта|Excel,Stata,Teamwork|8",
        "Максим|Git,Docker,SQL|5",
        "Денис|Python,Git,ML,Time management|3",
        "Виктория|Excel,Communication,Teamwork|1",
        "Кирилл|Leadership,Communication,Teamwork|2",
    ]

    courses_data = [
        "Data analysis|Excel,SQL,Python,Stata",
        "Machine learning|Python,ML,Git,Docker",
        "Soft skills|Teamwork,Communication,Leadership,Time management",
    ]

    with open("employees.txt", "w", encoding="utf-8") as f:
        f.writelines(line + "\n" for line in employees_data)
    with open("courses.txt", "w", encoding="utf-8") as f:
        f.writelines(line + "\n" for line in courses_data)

        
    """Читаем данные из файлов"""
    with open("employees.txt", "r", encoding="utf-8") as f:
        employees_lines = [line.strip() for line in f if line.strip()]
    with open("courses.txt", "r", encoding="utf-8") as f:
        courses_lines = [line.strip() for line in f if line.strip()]

    """Преобразуем строки в словари"""
    employees = {}
    for line in employees_lines:
        assert "|" in line, "Ошибка в данных сотрудника"
        name, skills_str, motivation = line.split("|")
        employees[name] = {"skills": {s.strip().lower() for s in skills_str.split(",")},
                           "motivation": int(motivation)}

    courses = {}
    for line in courses_lines:
        assert "|" in line, "Ошибка в данных курса"
        title, skills_str = line.split("|")
        courses[title] = {s.strip().lower() for s in skills_str.split(",")}


    final_report = []
    for course, required_skills in courses.items():
        final_report.append(f"КУРС: {course.upper()}")
        matched = []

        def score(skills: set, motivation: int) -> float:
            nonlocal matched
            match = skills & required_skills
            metric = len(match) / len(required_skills)
            if metric >= min_match and motivation >= motivation_limit:
                matched.append((metric, motivation))
            return metric * 10 + motivation

        """Отбираем кандидатов для каждого курса"""
        selected = list(filter(
            lambda name: score(employees[name]["skills"], employees[name]["motivation"]) > 5,
            employees
        ))

        """Сортируем кандидатов по уровню мотивации"""
        selected_sorted = sorted(
            selected,
            key=lambda n: employees[n]["motivation"],
            reverse=True
        )

        if selected_sorted:
            for name in selected_sorted:
                skills = employees[name]["skills"]
                final_report.append(f"{name} \n  Навыки: {', '.join(skills)}.\n  Мотивация - {employees[name]['motivation']}.")
        else:
            final_report.append("Нет подходящих сотрудников")
        final_report.append("" * 50)

    """Сохраняем результат в файл"""
    with open("training_report.txt", "w", encoding="utf-8") as f:
        f.writelines(line + "\n" for line in final_report)
```
