# Задача 1 (0,5 баллов) 

Дана строка, в которой числа разделены точкой с запятой. Преобразуйте её в список целых чисел и выведите максимальное и минимальное числа.

_In: `input_string = "5;12;88;42;7"`_

_Out: `Максимум: 88; Минимум: 5`_


```python
# здесь напишите решение (ᵔ◡ᵔ)

# Исходные значения
input_string = "5;12;88;42;7"
num_list = [int(x) for x in input_string.split(';')]

# Расчет максимального и минимального чисел
print('Максимальное значение:', max(num_list))
print('Минимальное значение:', min(num_list))
```

# Задача 3 (0,5 баллов)

_In: `['python', 'is', 'awesome', 'and', 'so', 'is', 'code']`_

_Out: `['python', 'awesome', 'code']`_

```python
# здесь напишите решение (ᵔ◡ᵔ)

# Исходные значения
words = ['python', 'is', 'awesome', 'and', 'so', 'is', 'code']

# Создание нового списка со словами длиной >3 символов
word_list = [i for i in words if len(i) > 3]
print(word_list)
```

### Альтернативное решение с помощью цикла for

```python
# Исходные значения
words = ['python', 'is', 'awesome', 'and', 'so', 'is', 'code']
words_list = []

# Создание нового списка со словами длиной >3 символов
for i in words:
    if len(i) > 3:
        words_list.append(i)

print(words_list)
```

# Задача 8 (1 балл)

Напишите программу, которая определяет рекомендуемый уровень вакансии на основе:

- Опыта работы кандидата (в годах);
- Наличия высшего образования (да/нет);
- Числа сертификатов об обучении (количество).

Критерии оценки:

- Intern: менее 1 года опыта, независимо от образования и сертификатов
- Junior: (1-2 года опыта + высшее образование) ИЛИ (1-3 года опыта + 2 и более сертификатов)
- Middle: (3-4 года опыта + высшее образование) ИЛИ (4-5 лет опыта + 3 и более сертификатов)
- Senior: (5-6 лет опыта + высшее образование) ИЛИ (6-8 лет опыта + 5 и более сертификатов)
- Lead: (более 6 лет опыта + высшее образование + 3 и более сертификата) ИЛИ (более 8 лет опыта + 10 и более сертификатов)
- Not_defined: все остальные случаи

```python
# здесь напишите решение (ᵔ◡ᵔ)

# Исходные значения
experience = int(input('Опыт работы: '))
education = input('Наличие высшего образования: ').lower().strip()
certificates = int(input('Количество сертификатов об обучении: '))

# Определение рекомендуемого уровня вакансии
if experience < 1:
    level = 'Intern'
elif (1 <= experience <= 2 and education == 'да') or (1 <= experience <= 3 and certificates >= 2):
    level = 'Junior'
elif (3 <= experience <= 4 and education == 'да') or (4 <= experience <= 5 and certificates >= 3):
    level = 'Middle'
elif (5 <= experience <= 6 and education == 'да') or (6 <= experience <= 8 and certificates >= 5):
    level = 'Senior'
elif (experience > 6 and education == 'да' and certificates >= 3) or (experience > 8 and certificates >= 10):
    level = 'Lead'
else:
    level = 'Not_defined'

print(f'\nРекомендуемый уровень вакансии: {level}')
```

# Задача 9 (1 балл)

Напишите игру **“HR угадывает зарплатные ожидания”**

Сюжет:
-  Кандидат хочет некоторую зарплату, но HR не знает какую (случайно сгенерированное число в диапазоне от 80 до 150, измеряется в тысячах рублей).
-  HR делает предложения:
      + Если предложение равно ожиданиям → кандидат принимает оффер, игра завершается.
      + Если предложение слишком низкое → кандидат недоволен и HR предлагает большую сумму.
      + Если предложение слишком высокое → кандидат пугается и HR предлагает меньшую сумму.
-  HR может попытаться максимум 7 раз. Если за это время не угадал → кандидат уходит.
-  Если HR вводит 0, значит он сдаётся раньше времени и игра завершается.

```python
# здесь напишите решение (ᵔ◡ᵔ)

from random import randint
salary = randint(80, 150)

for i in range(7):
    guess = int(input('Ваше предположение: '))
    if guess == 0:
        print('Конец игры!')
        break
    elif guess == salary:
        print(f'Кандидат принял оффер!')
        break
    elif guess < salary:
        print('Кандидат недоволен (предложите сумму больше) \n')
    else:
        print('Кандидат испугался (предложите сумму меньше)\n')
else:
    print(f'''Кандидат ушел! 
Зарплатные ожидания кандидата - {salary} тысяч рублей.''')
```

# Задача 10 (2 балла)

У вас есть данные о кандидатах на должность аналитика с их оценками по разным навыкам по 100-балльной шкале. Нужно проанализировать эти данные. 

Данные в формате списка кортежей `candidates`. Каждый кортеж содержит четыре значения: 
- имя_кандидата, 
- оценка_python, 
- оценка_sql, 
- оценка_коммуникации.

1. Выведите средние оценки по каждому навыку.
2. Выведите имена кандидатов, у которых все оценки выше 80.
3. Выведите имена кандидатов, у которых оценка по SQL выше 85.
4. Выведите имя кандидата с наивысшим средним баллом.
5. Выведите таблицу результатов с рейтингом кандидатов (чем выше сумма балов тем выше позиция в рейтинге).

```python
# здесь напишите решение (ᵔ◡ᵔ)

# Исходные значения
candidates = [
    ("Анна", 85, 92, 88),
    ("Борис", 78, 95, 82),
    ("Виктория", 92, 88, 95),
    ("Григорий", 76, 84, 79),
    ("Дарья", 89, 91, 87),
    ("Евгений", 81, 96, 83)
]
```

```python
# 1. Средние оценки по каждому навыку 
python_skill = sum(python for name, python, sql, communication in candidates) / len(candidates)
print(f'Средняя оценка по навыку Python: {python_skill:.1f}')

sql_skill = sum(sql for name, python, sql, communication in candidates) / len(candidates)
print(f'Средняя оценка по навыку SQL: {sql_skill:.1f}')

communication_skill = sum(communication for name, python, sql, communication in candidates) / len(candidates)
print(f'Средняя оценка по навыку Коммуникации: {communication_skill:.1f}')


# 2. Кандидаты, у которых все оценки выше 80
all_names = [name for name, python, sql, communication in candidates if python > 80 and sql > 80 and communication > 80]
print(f'\nКандидаты с оценками выше 80: {", ".join(all_names)}.')


# 3. Кандидаты, у которых оценка по SQL выше 85
sql_names = [name for name, python, sql, communication in candidates if sql > 85]
print(f'\nКандидаты с оценками по SQL выше 85: {", ".join(sql_names)}.')


# 4. Кандидат с наивысшим средним баллом
best_candidate = None
best_avg_score = 0 

for name, python, sql, communication in candidates:
    avg_score = (python + sql + communication) / 3
    if avg_score > best_avg_score:
        best_avg_score = avg_score
        best_candidate = name

print(f'\nКандидат с наивысшим средним баллом: {best_candidate} ({best_avg_score:.1f}).')


# 5. Таблица результатов с рейтингом кандидатов
ranked_candidates = []
for name, python, sql, communication in candidates:
    total = python + sql + communication
    ranked_candidates.append((name, python, sql, communication, total))

for i in range(len(ranked_candidates) - 1):
    for j in range(i + 1, len(ranked_candidates)):
        if ranked_candidates[i][4] < ranked_candidates[j][4]:
            ranked_candidates[i], ranked_candidates[j] = ranked_candidates[j], ranked_candidates[i]


print('\nРейтинговая таблица')
print('-' * 65)
print(f'''{'Место':<6} {'Имя':<12} {'Python':<8} {'SQL':<8} {'Коммуникации':<14} {'Сумма':<8}''')
print('-' * 65)

for i, (name, python, sql, communication, total) in enumerate(ranked_candidates, 1):
    print(f'{i:<6} {name:<12} {python:<8} {sql:<8} {communication:<14} {total:<8}')
```

## Дополнительное задание


HR-департамент корпорации "Сидней" провел ежегодный опрос удовлетворённости своих сотрудников. Каждому сотруднику было предложено оценить компанию по нескольким критериям:
- Условия труда 
- Уровень зарплаты 
- Взаимоотношения в коллективе 
- Лояльность компании 

Каждый критерий оценивался по шкале от 0 до 10, где 0 - совсем неудовлетворен, а 10 - максимально удовлетворен. Цель опроса - выявить узкие места в управлении, а также определить потенциальных сотрудников, которые могли бы покинуть корпорацию из-за имеющихся недовольств. 

HR собрал данные в виде списка кортежей: `(Имя, отдел, [оценка_1, оценка_2, оценка_3, оценка_4])`. Аналитику корпорации "Сидней" было поручено обработать собранные данные и подготовить аналитический отчёт для оперативного совещания. 

Для подготовки отчета аналитику необходимо: 
1. Посчитать среднюю удовлетворённость каждого сотрудника.
2. Разделить сотрудников на категории:
- “Лояльный” — средняя оценка больше 8
- “Нейтральный” — средняя оценка между 5 и 8
- “Рискует уйти” — средняя оценка меньше 5

3. Найти среднюю оценку по каждому отделу.
4. Определить самый “счастливый” и “недовольный” департамент.
5. Показать рейтинг сотрудников по уровню удовлетворенности.
6. Подсчитать количество сотрудников в каждой категории удовлетворенности.

```python
# Исходные значения
employees = [
    ("Анна", "IT", [9, 10, 8, 9]),
    ("Борис", "HR", [6, 7, 6, 5]),
    ("Виктория", "Маркетинг", [4, 5, 4, 3]),
    ("Глеб", "IT", [8, 8, 9, 10]),
    ("Дарья", "Продажи", [7, 6, 8, 7]),
    ("Елена", "HR", [9, 9, 10, 10]),
    ("Зоя", "Маркетинг", [3, 4, 4, 2]),
    ("Игорь", "Продажи", [8, 8, 7, 9]),
    ("Кирилл", "IT", [5, 6, 7, 6]),
    ("Лилия", "HR", [4, 5, 3, 5])
]
```

```python
# 1. Средняя удовлетворенность каждого сотрудника
employees_with_avg = []
for name, dep, scores in employees:
    avg = round(sum(scores) / len(scores), 2)
    employees_with_avg.append([name, dep, scores, avg])


# 2. Категория удовлетворенности
employees_with_status = []
for emp in employees_with_avg:
    name, dep, scores, avg = emp
    if avg >= 8:
        status = 'Лояльный'
    elif avg >= 5:
        status = 'Нейтральный'
    else:
        status = 'Рискует уйти'
    employees_with_status.append([name, dep, scores, avg, status])


# 3. Средняя удовлетворенность по отделам
departments = []
for emp in employees_with_status:
    dep = emp[1]
    if dep not in departments:
        departments.append(dep)

dept_avg = []
for dep in departments:
    dep_scores = []
    for emp in employees_with_status:
        if emp[1] == dep:
            dep_scores.append(emp[3])
    avg_score = round(sum(dep_scores) / len(dep_scores), 2)
    dept_avg.append([dep, avg_score])


# 4. Рейтинг департаментов
for i in range(len(dept_avg) - 1):
    for j in range(i + 1, len(dept_avg)):
        if dept_avg[i][1] < dept_avg[j][1]:
            dept_avg[i], dept_avg[j] = dept_avg[j], dept_avg[i]

best_dept = dept_avg[0]
worst_dept = dept_avg[-1]


# 5. Рейтинг сотрудников по среднему баллу
for i in range(len(employees_with_status) - 1):
    for j in range(i + 1, len(employees_with_status)):
        if employees_with_status[i][3] < employees_with_status[j][3]:
            employees_with_status[i], employees_with_status[j] = employees_with_status[j], employees_with_status[i]


# 6. Рейтинговая таблица
print('\nРейтинговая таблица')
print('-' * 75)
print(f'''{'Место':<6} {'Имя':<12} {'Отдел':<12} {'Средняя оценка':<15} {'Статус':<15}''')
print('-' * 75)

for i, (name, dep, scores, avg, status) in enumerate(employees_with_status, 1):
    print(f"{i:<6} {name:<12} {dep:<12} {avg:<15.1f} {status:<15}")


# 7. Средняя удовлетворенность по отделам
print('\nСредняя удовлетворенность по отделам:')
for dep, avg in dept_avg:
    print(f"{dep} — {avg:.1f}")

print(f'\nСамый довольный отдел: {best_dept[0]} ({best_dept[1]})')
print(f'Самый недовольный отдел: {worst_dept[0]} ({worst_dept[1]})')


# 8. Количество сотрудников по категориям
statuses = [emp[4] for emp in employees_with_status]
print('\nКоличество сотрудников по категориям:')
for cat in ['Лояльный', 'Нейтральный', 'Рискует уйти']:
    print(f"{cat}: {statuses.count(cat)}")


# 9. Список сотрудников, которые рискуют уйти
at_risk = [emp[0] for emp in employees_with_status if emp[4] == 'Рискует уйти']
print('\nСотрудники, требующие внимания HR:', ', '.join(at_risk) if at_risk else 'нет')
```
